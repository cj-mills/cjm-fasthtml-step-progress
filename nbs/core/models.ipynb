## StepInfo

```python
#| export
@dataclass
class StepInfo:
    """Describes a single step in a multi-step workflow."""
    title: str  # step name shown in tooltip on hover
```

```python
# Test StepInfo
step = StepInfo(title="Selection")
assert step.title == "Selection"

step2 = StepInfo("Segment & Align")
assert step2.title == "Segment & Align"

print("StepInfo tests passed!")
```
